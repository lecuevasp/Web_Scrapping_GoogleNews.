# Web Scraping  a Google News usando requests y BeautifulSoup

Para el presente anállisis, se busca la información en google noticias. Si se ingresa a la página de google, la url presenta la sigueinte estructura


`https://news.google.com/search?q=buscar%20informacion&hl=es-419&gl=CL&ceid=CL:es-419`

Se procede a segmentar dicha url para parametrizar la busqueda a realizar


```python
news_pag = 'https://news.google.com/'
parametro_busqueda = 'search?q='
busqueda_no_espacios = 'buscar%20informacion'
parametro_final = '&hl=es-419&gl=CL&ceid=CL:es-419'
```

A continuación mostraremos un ejemplo de extracción de información usando estas bibliotecas:

````python
import requests # Obtener html de página web
import pandas as pd # Manejo de datos
from bs4 import BeautifulSoup
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
````

````python
news_pag = "https://news.google.com/"
parametro_busqueda = "search?q="
busquedas = ['migracion chile', 'migracion irregular', 'campamento migrante', 'plaza migrante']
busqueda_no_espacios = [busqueda.replace(" ","%20") for busqueda in busquedas]
busqueda_no_espacios = '%2C'.join(busqueda_no_espacios)
parametro_final = "&hl=es-419&gl=CL&ceid=CL%3Aes-419"
html_dir = ''.join([news_pag, parametro_busqueda, busqueda_no_espacios, parametro_final])
html_dir
````

````python
url = requests.get(html_dir)
url_html = BeautifulSoup(url.content, parser = 'html')
````

## Generalizar la extracción por medio de una función noticias

Finalmente, procederemos a generalizar los pasos anteriores por medio de una función que extraiga las noticias desde google news.

````python
def fn_obtener_noticias(busqueda):
    '''
    Parámetros:
        busqueda(str): Debe ingresarse como una lista. En caso de que sea más de un concepto, se debe separa por comas

    Return:
        df_noticias(Frame): Retorna un DataFrame con la información extraida desde google news
    '''
    # Página de google news
    news_pag = "https://news.google.com/"
    # Pámetro de busqueda
    parametro_busqueda = "search?q="
    # Concepto a buscar
    busquedas = busqueda
    # Reemplazar espacios en blanco
    busqueda_no_espacios = [busqueda.replace(" ","%20") for busqueda in busquedas]
    # Unir términos y reemplazar comas
    busqueda_no_espacios = '%2C'.join(busqueda_no_espacios)
    # Parámetro final
    parametro_final = "&hl=es-419&gl=CL&ceid=CL%3Aes-419"
    html_dir = ''.join([news_pag, parametro_busqueda, busqueda_no_espacios, parametro_final])
    print(html_dir)
    url = requests.get(html_dir)
    url_html = BeautifulSoup(url.content, parser = 'html')
    df_noticias = pd.DataFrame()
    df_noticias['fechas'] = [pd.to_datetime(fecha['datetime'][:10])for fecha in url_html.find_all('time', {'class':'hvbAAd'})]
    df_noticias['fuentes'] = [fuente.get_text() for fuente in url_html.find_all('div', {'class':'a7P8l'})]
    df_noticias['noticias'] = [noticia.get_text() for noticia in url_html.find_all('a', {'class':'JtKRv'})]
    df_noticias['links'] = [''.join([news_pag, link['href'][2:]]) for link in url_html.find_all('a', {'class':'JtKRv'})]
    df_noticias = df_noticias[['fechas','fuentes','noticias','links']].drop_duplicates()
    df_noticias.sort_values(['fechas','noticias'], ascending=[False, True], inplace=True)
    return df_noticias
````

````python
df_noticias = fn_obtener_noticias(['politica', 'corrupcion'])
````
